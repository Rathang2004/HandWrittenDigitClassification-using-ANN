```python
#import the necessary libraries
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

```python
#load the handwritten datasets
(X_train, y_train) , (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
#i am just printing the shape so i can now flatten my data
print("This is X_train shape", X_train.shape)
print("This is y_train shape", y_train.shape)
print("This is X_test shape", X_test.shape)
print("This is y_test shape", y_test.shape)
```

```python
#reshape ur tensor into a 1-d array
X_train = X_train.reshape(-1,28*28)
X_test = X_test.reshape(-1,28*28)
print("This is X_train shape", X_train.shape)

#As you can see from printing this, that the images are from 0-255 range, we want to change them all to be between 0-1
#The purpose for doing that is to increase our accuracy, as the sigmoid function works from 0-1
print("This is X_train shape", X_train[59999])

X_train = X_train/255
X_test = X_test/255

#As you can see the same image is now a float value between 0 and 1
print("This is X_train shape",X_train[59999])
```

```python
#create your keras model
model = keras.Sequential([
    # for no hidden layers, just one input layer of the number of features 28*28 or the pixels and output layer of the 10 digits
    # sigmoid because we wanna get the max from 0-1

    #here we have 2 hidden layers and 4 layers in total: 784,256,128, and 10
    keras.layers.Dense(256, input_shape=(X_train.shape[1],), activation='relu'),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='sigmoid'),
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy'])

model.fit(X_train,y_train,epochs=20)
```

```python
# Note I ran into an error when I first did evaluate and my guess is because
# I didn't change my X_test, I should also flatten the array and change values on the 0-1 scale as well.
model.evaluate(X_test,y_test)
```
